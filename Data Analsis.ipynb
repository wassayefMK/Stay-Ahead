# Details about the dataset.

**CustomerID**: A unique ID that identifies each customer.\
**Gender**: The customer’s gender: Male, Female\
**Age**: The customer’s current age, in years.\
**Senior Citizen**: Indicates if the customer is 65 or older: Yes, No\
**Married (Partner)**: Indicates if the customer is married: Yes, No\
**Dependents**: Indicates if the customer lives with any dependents: Yes, No. Dependents could be children, parents, grandparents, etc.\
**Number of Dependents:** Indicates the number of dependents that live with the customer.\
**Phone Service**: Indicates if the customer subscribes to home phone service with the company: Yes, No\
**Multiple Lines**: Indicates if the customer subscribes to multiple telephone lines with the company: Yes, No\
**Internet Service**: Indicates if the customer subscribes to Internet service with the company: No, DSL, Fiber Optic, Cable.\
**Online Security**: Indicates if the customer subscribes to an additional online security service provided by the company: Yes, No\
**Online Backup**: Indicates if the customer subscribes to an additional online backup service provided by the company: Yes, No\
**Device Protection Plan**: Indicates if the customer subscribes to an additional device protection plan for their Internet equipment provided by the company: Yes, No\
**Premium Tech Support**: Indicates if the customer subscribes to an additional technical support plan from the company with reduced wait times: Yes, No\
**Streaming TV**: Indicates if the customer uses their Internet service to stream television programing from a third party provider: Yes, No. The company does not charge an additional fee for this service.\
**Streaming Movies**: Indicates if the customer uses their Internet service to stream movies from a third party provider: Yes, No. The company does not charge an additional fee for this service.\
**Contract**: Indicates the customer’s current contract type: Month-to-Month, One Year, Two Year.\
**Paperless Billing**: Indicates if the customer has chosen paperless billing: Yes, No\
**Payment Method**: Indicates how the customer pays their bill: Bank Withdrawal, Credit Card, Mailed Check\
**Monthly Charge**: Indicates the customer’s current total monthly charge for all their services from the company.\
**Total Charges**: Indicates the customer’s total charges, calculated to the end of the quarter specified above.\
**Tenure**: Indicates the total amount of months that the customer has been with the company.

**Churn**: Yes = the customer left the company this quarter. No = the customer remained with the company. Directly related to Churn Value.

# Data Cleaning

## Create a clean copy

```python
# Start from the original dataframe
df_clean = data.copy()
df_clean.head()
```

## Fix data type of TotalCharges

```python
# convert TotalCharges to numeric
tc_num = pd.to_numeric(df_clean["TotalCharges"], errors="coerce")
# number of rows with NaN in TotalCharges after conversion
tc_num.isna().sum()
```

* After attempting to convert TotalCharges to numeric, we obtain some NaN values.
* This means there are rows where TotalCharges is empty or invalid (e.g., blank strings), and we need to inspect and handle these rows.

```python
invalid_tc_rows = df_clean[tc_num.isna()][["customerID", "tenure", "MonthlyCharges", "TotalCharges"]]
invalid_tc_rows
```

* These rows have invalid or blank values in the TotalCharges column (they became NaN after conversion).
* Most of them have tenure = 0, which means these customers are extremely new and have not accumulated total charges yet.

## Drop rows with invalid TotalCharges

```python
# Convert TotalCharges to numeric (again, but now directly inside df_clean)
df_clean["TotalCharges"] = pd.to_numeric(df_clean["TotalCharges"], errors="coerce")

# Drop rows where TotalCharges is NaN
df_clean = df_clean.dropna(subset=["TotalCharges"]).reset_index(drop=True)

# Check new shape
df_clean.shape
```

* We converted the TotalCharges column to a numeric type and dropped all rows where TotalCharges was NaN.

## Re-check data types after cleaning

```python
df_clean.info()
```

* The TotalCharges column is now correctly stored as float64, and there are no missing values in this column.

## Check for duplicates

```python
# Number of duplicated full rows
df_clean.duplicated().sum()
```

* There are no duplicated rows in the cleaned dataset.

## Rename features

```python
rename_map = {
    "gender": "gender",
    "SeniorCitizen": "is_senior_user",
    "Partner": "has_partner",
    "Dependents": "has_dependents",
    "tenure": "subscription_tenure_months",
    "PhoneService": "ProductEnabled",
    "InternetService": "PlanType",
    "OnlineSecurity": "OnlineSecurity",
    "OnlineBackup": "OnlineBackup",
    "DeviceProtection": "DeviceProtection",
    "TechSupport": "TechSupport",
    "StreamingTV": "FeatureA",
    "StreamingMovies": "FeatureB",
    "Contract": "billing_cycle_type",
    "PaperlessBilling": "PaperlessBilling",
    "PaymentMethod": "PaymentMethod",
    "MonthlyCharges": "MonthlyCharges",
    "TotalCharges": "TotalCharges",
    "Churn": "Churn",
}

df_clean = df_clean.rename(columns=rename_map)
df_clean.head()
```

# Data Preprocessing

```python
df_model = df_clean.copy()
```

* We created a new dataframe called df_model as a modeling version of the data.
* This allows us to perform preprocessing steps (encoding, dropping columns, etc.) without affecting the cleaned dataset df_clean that we used for exploration.

## Encode Churn - 0 / 1

```python
df_model["Churn"] = df_model["Churn"].map({"No": 0, "Yes": 1})
df_model["Churn"].value_counts()
```

* We converted the Churn column from Yes/No to numeric values: 1 / 0
* This numeric encoding is required by most machine learning algorithms and clearly represents churn as a binary target variable.

## Drop customerID

```python
df_model = df_model.drop(columns=["customerID"])
df_model.columns
```

* We removed the customerID column because it is just an identifier and does not contain any meaningful information for predicting churn.
* Keeping it, may lead to overfitting, so it is better to exclude them from the modeling dataset.

## Separate features (X) and target (y)

```python
# Target
y = df_model["Churn"]

# Features
X = df_model.drop(columns=["Churn"])

X.head()
```

* This separation is a standard step before applying preprocessing and training machine learning models.

## Identify numeric vs categorical features

```python
# Numeric columns
num_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()

# Categorical columns (objects)
cat_cols = X.select_dtypes(include=["object"]).columns.tolist()

num_cols, cat_cols
```

* This distinction allows us to apply the appropriate preprocessing steps to each type of feature.

## One-Hot Encode categorical features

```python
# One-hot encode categorical columns
X_encoded = pd.get_dummies(X, columns=cat_cols, drop_first=True)

X_encoded.head()
```

* We applied one-hot encoding to all categorical features using pd.get_dummies.
* Each category was transformed into a separate binary (0/1) feature, and we used drop_first=True to avoid redundant dummy variables.

## Correlation matrix after one-hot encoding

```python
df_corr = X_encoded.copy()
df_corr["Churn"] = y

nunique = df_corr.nunique()
df_corr = df_corr.loc[:, nunique > 1]
df_corr = df_corr.loc[:, ~df_corr.T.duplicated()]
corr_matrix = df_corr.corr()

top_features = corr_matrix["Churn"].abs().sort_values(ascending=False).head(15).index
corr_top = corr_matrix.loc[top_features, top_features]



plt.figure(figsize=(12, 10))
sns.heatmap(
    corr_top,
    cmap="coolwarm",
    center=0,
    annot=True,
    fmt=".2f",
    square=True
)
plt.title("Correlation Matrix for Top Features Related to Churn")
plt.tight_layout()
plt.show()
```

Here is the a summary of the most related features to the target:
* Churn is negatively correlated with tenure, Contract_Two year, and TotalCharges, meaning long-term customers on two-year contracts with higher accumulated charges are less likely to churn.
* Churn is positively correlated with InternetService_Fiber optic, PaymentMethod_Electronic check, MonthlyCharges, and PaperlessBilling_Yes, suggesting that customers on expensive fiber plans who pay via electronic check and use paperless billing are more at risk of churn.
* There is a very strong correlation between tenure and TotalCharges, and also between MonthlyCharges and fiber optic service, indicating some redundancy / multicollinearity among billing-related features.

## Overall Summary

### Project Objective
The goal of this analysis is to support a **churn prediction system** for subscription (SaaS-like) products.  
We want to understand which factors drive customer churn so that the business can **estimate per-user churn probability** and **prioritize at-risk customers for retention offers** (discounts, plan changes, free extensions, etc.).

### Dataset
- Source: Telco Customer Churn dataset (`Churn.csv`).
- Each row represents one customer with information about **contract type, services, billing, and charges**.
- Target variable: **`Churn`** (`Yes`/`No`).

### Data Cleaning
- Converted `TotalCharges` from `object` to numeric.
- Removed **11 rows** with invalid/blank `TotalCharges`.
- Verified there are **no duplicate rows** and numeric columns have reasonable ranges.
- Dropped customerID

### Exploratory Analysis – Key Insights
- **Churn rate:** around **26–27%** of customers churn, so churn is a significant problem.
- **Tenure:** churned customers have **much lower tenure** than non-churned customers → new customers are more likely to leave.
- **MonthlyCharges:** churned customers pay **higher monthly charges** on average → higher prices are linked to higher churn.
- **Contract type:** month-to-month contracts show the **highest churn**, while one-year and two-year contracts have much lower churn rates.
- **Services & support:** customers **without** OnlineSecurity or TechSupport tend to churn more, suggesting that extra protection/support can help retention.
- **Payment method:** customers paying via **Electronic check** show higher churn compared to other payment methods.
- **Internet service:** **Fiber optic** customers have higher churn than DSL or customers with no internet service, indicating higher expectations or price sensitivity.

### Correlation Analysis
- `tenure` is **negatively correlated** with `Churn` (longer relationships → lower churn).
- `Contract_Two year` and higher `TotalCharges` are also negatively correlated with churn.
- `InternetService_Fiber optic` and `PaymentMethod_Electronic check` are **positively correlated** with churn and represent higher-risk segments.
- Customers with “no internet service” and lower `MonthlyCharges` tend to have **lower churn**, reflecting more stable, low-cost plans.

### Preprocessing for Modeling
- Encoded `Churn` as a binary variable (0 = No, 1 = Yes).
- Split the data into **features `X`** and **target `y`**.
- Applied **one-hot encoding** to all categorical features using `pd.get_dummies`, creating a fully numeric feature matrix `X_encoded`.
- Performed a **train/test split (80/20)** with stratification on `Churn`, preparing the data for training baseline models such as Logistic Regression and more advanced classifiers.

### Business Interpretation
Overall, the analysis shows that churn is driven mainly by:
- **Short tenure**, **high monthly charges**, **month-to-month contracts**, **fiber internet**, and **electronic check payments**.  
- These patterns can be used in a churn prediction system to:
  - Rank customers by churn risk.
  - Trigger targeted retention actions (e.g., moving high-risk users to longer contracts, adjusting prices, adding security/support features, or changing payment methods).
